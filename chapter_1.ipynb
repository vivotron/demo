###Linear regression using sklearn

```python
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model

#download and prepare data
dataurl="https://github.com/ageron/data/raw/main/"
data=pd.read_csv(dataurl+"lifesat/lifesat.csv")
```

```python
data.info()
```

```python
data.describe()
```

```python
data.head()
```

```python
x=data[['GDP per capita (USD)']].values  # x,y are 2d numpyarray of shape (n_samples,1)
y=data[['Life satisfaction']].values
```

```python
x.shape
```

```python
type(x)
```

```python
plt.scatter(x,y)
plt.title('Life satisfaction vs GDP per capita')
plt.xlabel('GDP per capita (USD)')
plt.ylabel('Life satisfaction')
plt.show()
```

```python
model=sklearn.linear_model.LinearRegression()
model.fit(x,y)
```

```python
model.predict([[37655.2]])
```
